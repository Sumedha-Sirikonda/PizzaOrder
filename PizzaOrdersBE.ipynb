Create Employee Class

```python

class Employee():
    
    def __init__(self,first_name,last_name):
        self.first_name= first_name
        self.last_name = last_name

    def set_first_name(self,first_name):
        self.first_name = first_name
    def set_last_name(self,last_name):
        self.last_name = last_name
        
    def get_first_name(self):
        return self.first_name
    def get_last_name(self):
        return self.last_name
    
    def __str__(self):
        return f"Employee Name:{self.get_first_name()} {self.get_last_name()}"
    
```

Store Class

```python
class Store(Employee):
    
    def __init__(self,address,zip_code,phone_no,mileage_no):

        self.address = address
        self.zip_code = zip_code
        self.phone_no = phone_no
        self.mileage_no = mileage_no
        self.emplist = []
        self.sales_dict = {}

#     setter methods
    def set_address(self,address):
        self.address = address
    def set_zip_code(self,zip_code):
        self.zip_code = zip_code
    def set_phone_number(self,phone_no):
        self.phone_no = phone_no
    def set_store_id(self,storeId):
        self.storeId = storeId
        
#     getter methods
    def get_address(self):
        return self.address
    def get_zip_code(self):
        return self.zip_code
    def get_phone_number(self):
        return self.phone_no
    
    
#     Add employees to the store list
    def add_employee(self,employee):
        self.emplist.append(Employee(employee.get_first_name(),employee.get_last_name()))
        
# Removes employee from the store list
    def remove_employee(self,employee):
        count = 0
        try:
            for i in self.emplist:
                if ((i.get_first_name() == employee.get_first_name()) and (i.get_last_name() == employee.get_last_name())):
                    count +=1
                    self.emplist.remove(i)
        
            if count == len(self.emplist):
                print("Employee doesn't work in the store")
        except:
            print('Employee doesnot exist')

    def set_pizza_sales(self,pizza_list,ddate):
#         for i in pizza_list:
        self.sales_dict[ddate] = pizza_list
        
        
    def monthly_pizza_sales(self,start_date,end_date):
        count = 0
        strt = datetime.strptime(start_date+' 00:00:00.00000', '%Y-%m-%d %H:%M:%S.%f')
        end = datetime.strptime(end_date+' 00:00:00.00000', '%Y-%m-%d %H:%M:%S.%f')
        for k,v in self.sales_dict.items():
            k = datetime.strptime(k, '%Y-%m-%d %H:%M:%S.%f')
            if strt<= k <= end:
                count += len(v)
        
        return count
         
         
    def __str__(self):
        return f"Store Address:{self.get_address()},Zip Code:{self.get_zip_code()}"
        
```

Pizza Class

```python
class Pizza():
    
    def __init__(self,pizza_name,toppings):
        self.pizza_name = pizza_name
        self.toppings = toppings
        
#         default crust type would be 'thick'
        self.crust_type = 'thick'
        self.topping_list = []
        
        self.price = 0
        self.specify_price()
        
    
    def set_pizza_name(self,pizza_name):
        self.pizza_name = pizza_name
    
    def get_pizza_name(self):
        return self.pizza_name
    
    def set_toppings(self,toppings):
        if len(toppings) >1:
            self.toppings = set(toppings)
        else:
            self.toppings = {}

    def get_toppings(self):
        return set(self.toppings)
    
    def get_topping_list(self):
        return self.topping_list
    
    def set_price(self,price):
        self.price = price
    
    def get_price(self):
        return self.price
    
    def list_of_toppings(self):
        return self.get_topping_list()
        
    def set_crust_type(self,crust_type):
        self.crust_type = crust_type
        
    def get_crust_type(self):
        return self.crust_type
    
    
#     user should be able to add extra toppings
    def add_toppings(self,topps):
        toppings = []
#         retrieve original toppings list
        toppings = self.get_toppings()
#         append the new toppings
        toppings.append(topps)
#         set it into toppings variable and topping_list variable
        for i in toppings:
            self.topping_list.append(i)
        
        self.set_toppings(toppings)
        self.specify_price()
        return self.topping_list
    
#     user should be able to remove toppings if not required
    def remove_toppings(self,tops):
#         retrieve original toppings list
        topps = tops
        toppings = self.get_toppings()
        for i in topps:
            if i in toppings:
                toppings.remove(i)
            else:
                print('Topping not listed: ',i)
        self.set_toppings(toppings)
        self.specify_price()
    
#     default crust type would be thick, if you want to change to thin call the method
    def specify_crust_type(self,crust):
        if crust.lower() == 'thick' or crust.lower() == 'thin':
            self.set_crust_type(crust)
        else:
            print('Only two crust types available')
    
    #gives the price of the pizza based on toppings and crust type
    def specify_price(self):
        price = 0
#         print(self.get_crust_type(),'  ',len(self.get_toppings()))
        if self.get_crust_type() == 'thick': 
            if len(self.get_toppings())<=3:
                price=400
            elif 4<=len(self.get_toppings())<=7:
                price = 600
            elif len(self.get_toppings())>=8:
                price = 900
        elif self.get_crust_type() == 'thin':
            
            if len(self.get_toppings())<= 3:
                price = 500
            elif 4<=len(self.get_toppings())<=7:
                price = 700
            elif len(self.get_toppings())>= 8:
                price = 1000
        
        total_price = price+self.get_price()
        self.set_price(total_price)
                
    def __str__(self):
        return f"Pizza Name: {self.get_pizza_name()}, Toppings:{self.get_toppings()},Crust Type: {self.get_crust_type()},Pizza amount: {self.get_price()}"
    
```

Create a PizzaOrder Class

```python
class PizzaOrder(Pizza):
    
    def __init__(self,promo_code,customer):
        self.promotion_code = promo_code
        self.order_status = ''
        self.customer = customer
        self.store = None
        self.pizza_list = []
        self.date_time = ''
        self.promo_code_list = ['FIRST20','COVID50','START15','LUCKY10','DUSBUS25','MBUFF60']
        
    def set_promotion_code(self,promotion_code):
        self.promotion_code =  promotion_code
    def set_order_status(self,order_status):
        self.order_status = order_status
    def set_store(self,store):
        self.store = store
    def set_date_time(self,date_time):
        self.date_time = date_time
    def set_customer(self,customer):
        self.customer = customer
    
    def get_promotion_code(self):
        return self.promotion_code
    def get_order_status(self):
        return self.order_status
    def get_store(self):
        return self.store
    def get_date_time(self):
        return self.date_time
    def get_customer(self):
        return self.customer
    
#     def get_pizza_list(self):
#         for i in self.pizza_list:
#             print(i)
    
    def gen_datetime(self):
    # generate a datetime in format yyyy-mm-dd hh:mm:ss.000000
        final_date = datetime.now()
#         print(final_date)
#         start = datetime(min_year, 1, 1, 00, 00, 00)
#         years = max_year - min_year + 1
#         end = start + timedelta(days=365 * years)
#         return start + (end - start) * random.random()
        return final_date
#     price after applying discount
    def get_total_order_price(self):
        total_price = 0

        for i in self.pizza_list:
            total_price += int(i.get_price())
        
        pcode = self.get_promotion_code()
        
        if pcode != '':
            if pcode.upper() == 'FIRST20':
                total_price = 0.8*total_price
            elif pcode.upper() == 'COVID50':
                total_price = 0.5*total_price
            elif pcode.upper() == 'START15':
                total_price = 0.85*total_price
            elif pcode.upper() == 'LUCKY10':
                total_price = 0.2*total_price
            elif pcode.upper() == 'DUSBUS25':
                total_price = 0.75*total_price
            elif pcode.upper() == 'MBUFF60':
                total_price = 0.6*total_price
        
        return total_price
        
#      Ability to add pizza to the order   
    def add_pizza(self,pizza):
#         Adding pizza to the list
#         pizza object can hold from one to 5 number of pizzas, since a pizza order can have multiple pizzas
        for i in pizza:
            self.pizza_list.append(Pizza(i.get_pizza_name(),i.get_toppings()))
                                   
#         when pizza added to list order status to be created 
        self.set_order_status('ORDER_CREATED')
#         order date time would be time of order creation
        self.set_date_time(str(self.gen_datetime()))
        
        self.set_price(self.get_total_order_price())
        return self.pizza_list
        
#        Ability to remove pizza from the order 
    def remove_pizza(self,pizza_removed):

        for pizza in pizza_removed:
            if pizza in self.pizza_list:
                self.pizza_list.remove(pizza)
            else:
                print('Pizza not in order: ',pizza.get_pizza_name())
        if len(self.pizza_list)==0:
            self.set_order_status('ORDER_CANCELLED')
        
        self.set_price(self.get_total_order_price())
        return self.pizza_list

    
#     Ability to specify store from which the order is made
    def store_ordered(self,store):

        store.set_pizza_sales(self.pizza_list,self.get_date_time())
        self.set_store(store)
        
#     Retrieves the status of order based on time of order    
    def check_order_status(self):
#         assuming the order gets delivered in 1 hour
        status = self.get_order_status()
        time = self.get_date_time()
        t1 = datetime.strptime(time,'%Y-%m-%d %H:%M:%S.%f')
#         should check order status based on now time
        now = datetime.now()
        time_diff = (now - t1).seconds/60

        #         if order status not ORDER_CANCELLED
        if status == 'ORDER_CANCELLED':
            return self.get_order_status()
        else:
            #         If order creation time is > 2hr return ORDER_COMPLETE
            if time_diff > 120:
                self.set_order_status('ORDER_COMPLETE')
                return self.get_order_status()
            #         If order creation time is > 45 min & <1hr return ORDER_ON_DELIVERY
            elif 45 <time_diff < 60:
                self.set_order_status('ORDER_ON_DELIVERY')
                return self.get_order_status()
            #         If order creation time is > 30 mins & <45 mins return ORDER_READY
            elif 30 <time_diff < 45:
                self.set_order_status('ORDER_READY')
                return self.get_order_status()
            #         If order creation time is < 30 mins return ORDER_IN_PROCESS
            elif time_diff < 30:
                self.set_order_status('ORDER_IN_PROCESS')
                return self.get_order_status()
        
#     Ability to hold customer details
    def customer_info(self,customer):

        self.set_customer(customer)
        return self.get_customer()
    
    def __lt__(self,other):
        if self.get_date_time() == other.get_date_time():
            self.get_total_order_price() < other.get_total_order_price()
        return self.get_total_order_price() < other.get_total_order_price()
    
    def __str__(self):
        return f"Order Amount:{self.get_total_order_price()},Order Date :{self.get_date_time()[:10]},Promo Applied:{self.get_promotion_code()}, Order Status:{self.get_order_status()} \nCustomer Name: {self.get_customer().get_first_name()} {self.get_customer().get_last_name()}, No. of pizzas ordered: {len(self.pizza_list)},Delivered from store:{self.get_store().get_address()}"
```
